```python
import zipfile
import pandas as pd
with zipfile.ZipFile("CHN.zip") as z2:
    with z2.open("API_CHN_DS2_en_csv_v2.csv") as f2:
        df_CHN = pd.read_csv(f2, header=0, delimiter=",", skiprows=4)
        df_CHN = df_CHN.T
        df_CHN.columns = df_CHN.iloc[2]
        df_CHN = df_CHN.iloc[4:]
        df_CHN = df_CHN[['GDP (current US$)','CO2 emissions (metric tons per capita)','Population, total','Life expectancy at birth, total (years)','GNI (current US$)','Time required to start a business (days)','School enrollment, primary (% gross)','Foreign direct investment, net inflows (BoP, current US$)','Improved sanitation facilities (% of population with access)']]
        renaming_dict = { 
            'GDP (current US$)': 'GDP', 
            'CO2 emissions (metric tons per capita)': 'CO2_emissions', 
            'Population, total': 'Population_total',
            'Life expectancy at birth, total (years)': 'Life_expectancy',
            'GNI (current US$)': 'GNI',
            'Time required to start a business (days)': 'Time_required_to_start_a_business',
            'School enrollment, primary (% gross)': 'School_enrollment',
            'Foreign direct investment, net inflows (BoP, current US$)': 'Net_Foreign_Direct_Investment',
            'Improved sanitation facilities (% of population with access)': 'Improved_Sanitation_Facilities'
        }
        df_CHN.rename(columns=renaming_dict, inplace=True)
        df_CHN = df_CHN.drop(['Unnamed: 62'])
        df_CHN['Year']=df_CHN.index
        
        df_CHN.reset_index(drop=True,inplace=True)
        df_CHN['Country_Name']='China'
        
        #df_CHN = df_CHN.set_index('Year')
        #df_CHN.rename(columns={'Indicator Name': 'A'})
df_CHN
```

```python
df_CHN["GDP"] = pd.to_numeric(df_CHN["GDP"])
df_CHN["CO2_emissions"] = pd.to_numeric(df_CHN["CO2_emissions"])
df_CHN["Population_total"] = pd.to_numeric(df_CHN["Population_total"])
df_CHN["Life_expectancy"] = pd.to_numeric(df_CHN["Life_expectancy"])
df_CHN["GNI"] = pd.to_numeric(df_CHN["GNI"])
df_CHN["Time_required_to_start_a_business"] = pd.to_numeric(df_CHN["Time_required_to_start_a_business"])
df_CHN["School_enrollment"] = pd.to_numeric(df_CHN["School_enrollment"])
df_CHN["Net_Foreign_Direct_Investment"] = pd.to_numeric(df_CHN["Net_Foreign_Direct_Investment"])
df_CHN["Improved_Sanitation_Facilities"] = pd.to_numeric(df_CHN["Improved_Sanitation_Facilities"])
df_CHN['Year'] = pd.to_datetime(df_CHN['Year'], format='%Y')
df_CHN['Country_Name'] = pd.Categorical(df_CHN['Country_Name'])
#df_CHN[df_CHN.columns] = df_CHN[df_CHN.columns].apply(pd.to_numeric, errors='coerce')
```

Query 1: Time needed to start a business varying counbtry-by-country for the last 5 years (2013 -2017)

```python
Bus_Start_v_USA = pd.read_sql('''SELECT US.Year, sum(US.Time_required_to_start_a_business) AS US_Business_Start_Time, 
sum(B.Time_required_to_start_a_business) AS Bhutan_Business_Start_Time, sum(C.Time_required_to_start_a_business) AS China_Business_Start_Time, 
sum(G.Time_required_to_start_a_business) AS Germany_Business_Start_Time, sum(A.Time_required_to_start_a_business) AS Argentina_Business_Start_Time, 
sum(J.Time_required_to_start_a_business) AS Japan_Business_Start_Time, sum(R.Time_required_to_start_a_business) AS Russia_Business_Start_Time, 
sum(GH.Time_required_to_start_a_business) AS Ghana_Business_Start_Time
FROM Argentina A inner join Bhutan B ON A.Year=B.Year 
inner join China C on C.Year=A.Year
inner join Germany G ON G.Year = A.Year
inner join Japan J ON J.Year = A.Year
inner join Russia R ON R.Year=A.Year
inner join USA US ON US.Year=A.Year
inner join Ghana GH ON GH.Year=A.Year
Group By US.Year;''',con=engine)
Bus_Start_v_USA.iloc[-5:].set_index('Year').plot(kind='bar', figsize=(30,10))
```

Query 2: C02 emissions per capita and GDP per capita plotted against each other for all the countries. Each point plotted is a year for which both metrics for the country are avaliable. 

```python
GDP_PER_CAPITA_CO2 = pd.read_sql('''SELECT u.Year, u.GDP / u.Population_total AS 'USA GDP PER CAPITA', u.CO2_emissions AS 'USA C02 EMISSIONS PER CAPITA', d.GDP / d.Population_total AS 'GERMANY GDP PER CAPITA', d.CO2_emissions AS 'GERMANY C02 EMISSIONS PER CAPITA', a.GDP / a.Population_total AS 'ARGENTINA GDP PER CAPITA', a.CO2_emissions AS 'ARGENTINA C02 EMISSIONS PER CAPITA', b.GDP / b.Population_total AS 'BHUTAN GDP PER CAPITA', b.CO2_emissions AS 'BHUTAN C02 EMISSIONS PER CAPITA', r.GDP / r.Population_total AS 'RUSSIA GDP PER CAPITA', r.CO2_emissions AS 'RUSSIA C02 EMISSIONS PER CAPITA', c.GDP / c.Population_total AS 'CHINA GDP PER CAPITA', c.CO2_emissions AS 'CHINA C02 EMISSIONS PER CAPITA', j.GDP / j.Population_total AS 'JAPAN GDP PER CAPITA', j.CO2_emissions AS 'JAPAN C02 EMISSIONS PER CAPITA', g.GDP / g.Population_total AS 'GHANA GDP PER CAPITA', g.CO2_emissions AS 'GHANA C02 EMISSIONS PER CAPITA'                    
FROM USA u 
JOIN Germany d 
ON u.Year = d.Year 
JOIN Argentina a 
ON u.Year = a.Year 
JOIN Bhutan b 
ON u.Year = b.Year 
JOIN Russia r 
ON u.Year = r.Year 
JOIN China c 
ON u.Year = c.Year 
JOIN Japan j 
ON u.Year = j.Year
JOIN Ghana g
ON u.Year = g.Year
''', con=engine)
```

```python
ax = GDP_PER_CAPITA_CO2.plot(kind='scatter', x='USA GDP PER CAPITA', y='USA C02 EMISSIONS PER CAPITA', color = 'Blue', label = 'USA')
GDP_PER_CAPITA_CO2.plot(kind='scatter', x='GERMANY GDP PER CAPITA', y='GERMANY C02 EMISSIONS PER CAPITA', color = 'Orange', label = 'Germany', ax=ax)
GDP_PER_CAPITA_CO2.plot(kind='scatter', x='ARGENTINA GDP PER CAPITA', y='ARGENTINA C02 EMISSIONS PER CAPITA', color = 'Red', label = 'Argentina', ax=ax)
GDP_PER_CAPITA_CO2.plot(kind='scatter', x='BHUTAN GDP PER CAPITA', y='BHUTAN C02 EMISSIONS PER CAPITA',color = 'Purple', label = 'Bhutan', ax=ax)
GDP_PER_CAPITA_CO2.plot(kind='scatter', x='CHINA GDP PER CAPITA', y='CHINA C02 EMISSIONS PER CAPITA', color = 'Pink', label = 'China', ax=ax)
GDP_PER_CAPITA_CO2.plot(kind='scatter', x='JAPAN GDP PER CAPITA', y='JAPAN C02 EMISSIONS PER CAPITA', color = 'Green', label = 'Japan', ax=ax)
GDP_PER_CAPITA_CO2.plot(kind='scatter', x='GHANA GDP PER CAPITA', y='GHANA C02 EMISSIONS PER CAPITA', color = 'Black', label = 'Ghana', ax=ax)
GDP_PER_CAPITA_CO2.plot(kind='scatter', x='RUSSIA GDP PER CAPITA', y='RUSSIA C02 EMISSIONS PER CAPITA', color = 'Indigo', label = 'Russia',ax=ax)
ax.set_xlabel("GDP per capita")
ax.set_ylabel("CO2 emissions per capita")
```
